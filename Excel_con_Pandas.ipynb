# Leyendo archivos Excel

## Importando archivos Excel

```python
import pandas as pd
```

```python
# Leyendo archivo
datos_co2 = pd.read_excel('/content/emisiones_CO2.xlsx')
datos_co2.head()
```

```python
# Examinando las hojas que contiene el Excel
pd.ExcelFile('/content/emisiones_CO2.xlsx').sheet_names
```

## Parámetros de la función read_excel
Documentación: https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html

```python
# Importando por hoja
percapita = pd.read_excel('/content/emisiones_CO2.xlsx', sheet_name='emisiones_percapita')
percapita.tail()
```

```python
# Importando por hoja
fuentes = pd.read_excel('/content/emisiones_CO2.xlsx', sheet_name='fuentes')
fuentes.sample(5)
```

```python
# Seleccionando hoja y columnas a importar
intervalo = pd.read_excel('/content/emisiones_CO2.xlsx', sheet_name='emisiones_C02', usecols='A:D')
intervalo.head()
```

```python
# Seleccionando hoja, columnas y filas a importar
intervalo_2 = pd.read_excel('/content/emisiones_CO2.xlsx', sheet_name='emisiones_C02', usecols='A:D', nrows=10)
intervalo_2
```

## Escribiendo archivos Excel

```python
intervalo.to_excel('co2_percapita.xlsx', index=False)
```

```python
# Abriendo archivo para verificar los datos
pd.read_excel('co2_percapita.xlsx')
```
