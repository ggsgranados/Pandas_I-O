# Leyendo banco de datos

## Creando un banco de datos local

```python
# ORM a utilizar
import sqlalchemy
```

```python
# Importaciones:
# create_engine genera el entorno de la base de datos
# MetaData guarda la informacion
# Table para poder trabajar con ellas
# inspect para inspeccionar bancos de datos
# text para convertir las querys en texto
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
# Usando la memoria local de Collab
engine = create_engine('sqlite:///:memory:')
```

## Escribiendo en un banco de datos

```python
import pandas as pd
```

```python
archivo = '/content/clientes_banco.csv'
```

```python
# Colocando el csv en un dataframe para poder ingresarlo en la base de datos
datos = pd.read_csv(archivo)
```

```python
datos.head(4)
```

```python
# Parámetros
# (nombre, motor, quitando_indices)
datos.to_sql('clientes', engine, index=False)
```

```python
# Revisando lo que hay en el motor y asignandolo en una variable
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

## Leyendo una consulta SQL

```python
# Almacenando la query
query = 'SELECT * FROM clientes WHERE Categoria_de_renta = "Empleado"'
```

```python
# Forma 01 de realizar la query
empleados = pd.read_sql(sql=text(query), con=engine.connect())
```

```python
# Forma 02 de realizar la query
empleados2 = pd.read_sql(query, engine)
```

```python
empleados
```

```python
type(empleados)
```

```python
# Guardando la nueva tabla
empleados.to_sql('empleados', con=engine.connect(), index=False)
```

```python
pd.read_sql_table('empleados', engine)
```

```python
# Filtrando resultados
pd.read_sql_table('empleados', engine, columns=['ID_Cliente', 'Grado_estudio','Rendimiento_anual'])
```

## Actualizando un banco de datos

```python
query = 'Select * from clientes'
```

```python
pd.read_sql(sql=text(query), con=engine.connect())
```

```python
# Borrando registros
# Manejo de errores
from sqlalchemy.exc import SQLAlchemyError

query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print('#Registros borrados', r_set.rowcount)
```

```python
pd.read_sql_table('clientes', engine)
```

```python
# Actualizando registros
# Manejo de errores

query = 'UPDATE clientes SET Grado_estudio = "Nivel superior" WHERE ID_Cliente = 5008808'
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print('#Registros actualizados', r_set.rowcount)
```

```python
pd.read_sql_table('clientes', engine)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```
