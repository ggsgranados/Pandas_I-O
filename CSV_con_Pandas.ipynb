## Importando archivos CSV

## Parámetros de la función read_csv
Documentación: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

```python
# Leyendo archivo separado por punto y coma, indicándolo en el parámetro
datos_punto_coma = pd.read_csv('/content/superstore_data_punto_coma.csv',sep=';')
datos_punto_coma.head()
```

```python
# Usando nrows para leer una cantidad determinada de filas
datos_primeras_lineas = pd.read_csv('/content/superstore_data.csv', nrows=5)
datos_primeras_lineas
```

```python
# Usando usecols para leer columnas de interes por nombres
datos_columnas_especificas = pd.read_csv('/content/superstore_data.csv', usecols=['Id','Year_Birth','Income'])
datos_columnas_especificas.head()
```

```python
# Usando usecols para leer columnas de interes por posicion
datos_columnas_especificas = pd.read_csv('/content/superstore_data.csv', usecols=[0,1,4])
datos_columnas_especificas.head()
```

## Escribiendo archivos CSV
Documentación: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html

```python
# Guardando el DataFrame creado en un CSV, sin el índice
datos_columnas_especificas.to_csv('clientes_mercado.csv', index=False)
```
